### Scrape and clean wikipedia for lists of tennis champions

```python
import pandas as pd
```

```python
# read html as data. grab table nr 5 to use names of players
# Male players first, further down also Female players
# lastly, make csv-file and save in data assets folder

url = "https://en.wikipedia.org/wiki/List_of_ATP_number_1_ranked_singles_tennis_players"
df_male_player = pd.read_html(url)[5]
```

```python
print(len(df_male_player))
```

```python
df_male_player
```

```python
male_names = df_male_player.iloc[1:, 1].dropna().tolist()
print(df_male_player.iloc[:, 1].head(10))
```

```python
sliced_male_player = df_male_player.iloc[1:, 1]
print(sliced_male_player)
```

```python
import re
raw_male_names = df_male_player.iloc[1:, 1].tolist()

male_names = []
for name in raw_male_names:
    name = re.sub(r"\s*\(.*?\)|\[.*?\]", "", name)
    name = name.replace("\xa0", " ").strip()
    if name and name not in male_names:
        male_names.append(name)

print(male_names[:10])
```

```python
pd.Series(male_names).to_csv("data/male_players.csv", index = False)
```
